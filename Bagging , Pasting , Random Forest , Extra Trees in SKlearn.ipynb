## **Imports**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.ensemble import *
from sklearn.tree import *
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler , StandardScaler

from sklearn.metrics import accuracy_score
```

## **Read Data**

```python
data = pd.read_csv('data/heart.csv')
data
```

## **Train Test Split**

```python
X = data.iloc[: , :-1]
y = data.iloc[: , -1]

X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.15, random_state=0 , stratify=y)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## **Normalization**

```python
scaler = MinMaxScaler()
columns = ['age' , 'trestbps' , 'chol' , 'thalach' , 'oldpeak']
X_train[columns] = scaler.fit_transform(X_train[columns])
```

```python
X_test[columns] = scaler.transform(X_test[columns])
```

## **Bagging Sklearn**

```python
base_estimator = DecisionTreeClassifier(max_depth=3 , class_weight='balanced', random_state=0 , ccp_alpha=0.01)
Bagging_model = BaggingClassifier(base_estimator=base_estimator , n_estimators=50 , random_state=0 , bootstrap=True , max_samples=70)
Bagging_model.fit(X_train , y_train)
```

```python
print(f'Score of Bagging model on Train data is {Bagging_model.score(X_train , y_train):.3f}')
print(f'Score of Bagging model on Test data is {Bagging_model.score(X_test , y_test):.3f}')
```

## **Pasting Sklearn**

```python
base_estimator = DecisionTreeClassifier(max_depth=3 , class_weight='balanced', random_state=0 , ccp_alpha=0.01)
pasting_model = BaggingClassifier(base_estimator=base_estimator , n_estimators=20 , random_state=0 , bootstrap=False , max_samples=50)
pasting_model.fit(X_train , y_train)
```

```python
print(f'Score of Pasting model on Train data is {pasting_model.score(X_train , y_train):.3f}')
print(f'Score of Pasting model on Test data is {pasting_model.score(X_test , y_test):.3f}')
```

## **Random Forest Sklearn**

```python
rf = RandomForestClassifier(n_estimators=20 , max_depth=3 , class_weight='balanced' , random_state=0 , max_samples=50 , max_features=2)
rf.fit(X_train , y_train)
```

```python
print(f'Score of Random Forest model on Train data is {rf.score(X_train , y_train):.3f}')
print(f'Score of Random Forest on Test data is {rf.score(X_test , y_test):.3f}')
```

```python
train_scores = []
test_score = []
for i in range(1 , len(data.columns)):
    rf = RandomForestClassifier(n_estimators=20 , max_depth=3 , class_weight='balanced' , random_state=0 , max_samples=50 , max_features=i)
    rf.fit(X_train , y_train)
    train_scores.append(rf.score(X_train , y_train))
    test_score.append(rf.score(X_test , y_test))
```

```python
plt.figure(figsize=(7 , 5))
sns.lineplot(train_scores , c='b' , label='train score');
sns.lineplot(test_score , c='r' , label='test score');
plt.xlabel('Number Of Features')
plt.legend();
```

## **Extra Trees Sklearn**

```python
extra_classifier = ExtraTreesClassifier(n_estimators=20 , max_depth=3 , random_state=0 , class_weight='balanced' , max_samples=0.5 , bootstrap=True , max_features=7)
extra_classifier.fit(X_train , y_train)
```

```python
print(f'Score of Extra Trees model on Train data is {extra_classifier.score(X_train , y_train):.3f}')
print(f'Score of Extra Trees on Test data is {extra_classifier.score(X_test , y_test):.3f}')
```
